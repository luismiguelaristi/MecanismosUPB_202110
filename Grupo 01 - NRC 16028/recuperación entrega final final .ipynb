## How it works?
A quick return mechanism is used to convert rotary motion into reciprocating motion, unlike a normal slider crank mechanism, the forward reciprocating motion is slower than the return stoke

## Who made the quick retourn machanism
the autor or the quick retourn mechanism whas Joseph Whitworth, he made this mechanism in the mid-1800s.Using kinematics, he determined that the force and geometry of the rotating joint would affect the force and motion of the connected arm.

## About the mechanism

QR --> Quick Return

QR mechanisms feature different input durations for their working and return
strokes. The time ratio (TR) of a QR mechanism is the ratio of the change in input displacement during
the working stroke to its change during the return stroke. Several basic types of mechanism have a QR
action. These types include slider-crank and four-bar mechanisms. A project on QR mechanism design,
within a first course on the theory of mechanisms, has been found to be effective for exposing students
to concepts of mechanism design and analysis. This paper reviews basic QR mechanisms, presents a
project problem and solution examples, and discusses the value of inclusion of such project problems
within theory-of-mechanism courses.

Ron P. Podhorodeski, Scott B. Nokleby and Jonathan D. Wittchen. (Quick-return mechanism design and analysis projects). . Victoria, British Columbia, Canada. Department of Mechanical Engineering, University of Victoria

## Function

```python
import numpy as np
from scipy.optimize import fsolve 
import matplotlib.pyplot as plt

def qr(condIni, params):
#principal
    th2 = params[0]
    
#parameters
    r2 = params[1]
    r4 = params[2]
    r5 = params[3]
    r6y = params[4]
    th6 = params[5]
    th3y = params[6]
    th6y = params[7]
    r3y = params[8]
    
#initial conditions
    th3 = condIni[0]
    r3 = condIni[1]
    th5 = condIni[2]
    r6x = condIni[3]
    
    
    Y = np.zeros(4)
    Y[0] = r2*np.cos(th2) + r4*np.cos(th3) + r5*np.cos(th5) + r6x*np.cos(th6) - r6y*np.cos(np.pi/2);
    
    Y[1] = r2*np.sin(th2) + r4*np.sin(th3) + r5*np.sin(th5) + r6x*np.sin(th6) - r6y*np.sin(np.pi/2);
    
    Y[2] = -r2*np.cos(th2) + r3*np.cos(th3) + r3y*np.cos(np.deg2rad(270))
    
    Y[3] = -r2*np.sin(th2) + r3*np.sin(th3) + r3y*np.sin(np.deg2rad(270))
    
    return Y
    pass
```

## Velocity analysis: theory


speed analysis
We apply the following equation to model the speed of each rigid body (bar) of the mechanism:
$\overrightarrow{V_B} = \overrightarrow{V_A}+\overrightarrow{\omega}\times \overrightarrow{r_{AB}}$,
when  $\overrightarrow{\omega}$ is the angular velocity vector of the analyzed bar.

```python
import sympy as sym
from sympy.physics.mechanics import dynamicsymbols, init_vprinting

sym.init_printing(use_unicode=False, wrap_line=True)
init_vprinting()
```

```python
t = sym.Symbol('t')
# Parameters
r2s = sym.Symbol('r_2')
r5s = sym.Symbol('r_5')
r4s = sym.Symbol('r_4')
r6ys = sym.Symbol('r_6x')
r3ys = sym.Symbol('r_3y')
th6s = sym.Symbol('th_6')
th3ys = sym.Symbol('th_3y')
th6ys = sym.Symbol('th_6y')

# Variables defined as a function and then made dependent on t 
th2s, th3s, r3s, th5s, r6xs= sym.symbols('theta_2 theta_3 r_3 theta_5 r_6x', cls=sym.Function)

# First derivative of the variables with respect to time
th2sp = sym.diff(th2s(t),t)
th3sp = sym.diff(th3s(t),t)
r3sp = sym.diff(r3s(t),t)
th5sp = sym.diff(th5s(t),t)
r6xsp = sym.diff(r6xs(t),t)

# Second derivative of the variables with respect to time
th2spp = sym.diff(th2s(t),t,t)
th3spp = sym.diff(th3s(t),t,t)
r3spp = sym.diff(r3s(t),t,t)
th5spp = sym.diff(th5s(t),t,t)
r6xspp = sym.diff(r6xs(t),t,t)
```

```python
eq1 = r2s*sym.cos(th2s(t)) + r4s*sym.cos(th3s(t)) + r5s*sym.cos(th5s(t)) + r6xs(t)*sym.cos(th6)- r6ys*sym.cos(sym.pi/2);
eq2 = r2s*sym.sin(th2s(t)) + r4s*sym.sin(th3s(t)) + r5s*sym.sin(th5s(t)) + r6xs(t)*sym.sin(th6)- r6ys*sym.sin(sym.pi/2);
eq3 = -r2s*sym.cos(th2s(t)) + r3s(t)*sym.cos(th3s(t)) + r3ys*sym.cos(sym.pi*3/2);
eq4 = -r2s*sym.sin(th2s(t)) + r3s(t)*sym.sin(th3s(t)) + r3ys*sym.sin(sym.pi*3/2);

eq1,eq2,eq3,eq4
```

Ahora las derivamos con respecto al tiempo

```python
eq1p = sym.diff(eq1,t)
eq2p = sym.diff(eq2,t)
eq3p = sym.diff(eq3,t)
eq4p = sym.diff(eq4,t)
eq1p
```

#### Expressing the system of equations in matrix form

To do this we must use the function "linear_eq_to_matrix", the problem is that this function does not recognize the derivatives as the unknowns, we must replace the derivatives by symbols to generate the matrix system.

We can see that these symbols are identical to the derivatives with respect to time, but by defining them in this way, for sympy they are not related to time (t).

Now we only need to replace using the "subs" function of sympy.

```python
th2ssym, th3ssym, r3ssym, th5ssym, r6xssym = sym.symbols('theta_2 theta_3 r_3 theta_5 r_6x')
th2spsym, th3spsym, r3spsym, th5spsym, r6xspsym = sym.symbols('\dot{\\theta_2} \dot{\\theta_3} \dot{\\r_3} \dot{\\theta_5} \dot{\\r_6x}')
th2sppsym, th3sppsym, r3sppsym, th5sppsym, r6xsppsym = sym.symbols('\ddot{\\theta_2} \ddot{\\theta_3} \ddot{\\r_3} \ddot{\\theta_5} \ddot{\\r_6x}')
th2ssym,th2spsym,th2sppsym,
```

```python
eq1psym = eq1p.subs([
                    (th2sp,th2spsym),
                    (th3sp,th3spsym),
                    (r3sp,r3spsym),
                    (th5sp,th5spsym),
                    (r6xsp,r6xspsym)
                    ])
eq2psym = eq2p.subs([
                    (th2sp,th2spsym),
                    (th3sp,th3spsym),
                    (r3sp,r3spsym),
                    (th5sp,th5spsym),
                    (r6xsp,r6xspsym)
                    ])
eq3psym = eq3p.subs([
                    (th2sp,th2spsym),
                    (th3sp,th3spsym),
                    (r3sp,r3spsym),
                    (th5sp,th5spsym),
                    (r6xsp,r6xspsym)
                    ])
eq4psym = eq4p.subs([
                    (th2sp,th2spsym),
                    (th3sp,th3spsym),
                    (r3sp,r3spsym),
                    (th5sp,th5spsym),
                    (r6xsp,r6xspsym)
                    ])
eqns = [eq1psym,eq2psym,eq3psym,eq4psym]
As, bs = sym.linear_eq_to_matrix(eqns, [th3spsym,r3spsym,th5spsym,r6xspsym])
As
```

```python
# Analytical speed
th3pv=np.zeros(numpos)
r3pv=np.zeros(numpos)
th5pv=np.zeros(numpos)
r6xpv=np.zeros(numpos)

# Analytical Acceleration
th3ppv=np.zeros(numpos)
r3ppv=np.zeros(numpos)
th5ppv=np.zeros(numpos)
r6xppv=np.zeros(numpos)

velAm=np.zeros((numpos,3))
velBm=np.zeros((numpos,3))
velCm=np.zeros((numpos,3))


for i in range(0,numpos):
    # Speed
    
    am = [[-r4*np.sin(th3v[i]),0,-r5*np.sin(th5v[i]),-1.0],
     [r4*np.cos(th3v[i]),0,r5*np.cos(th5v[i]),0],
     [-r3v[i]*np.sin(th3v[i]),np.cos(th3v[i]),0,0],
     [r3v[i]*np.cos(th3v[i]),np.sin(th3v[i]),0,0]]
    
    bv = [r2*np.sin(th2v[i])*th2p,
     -r2*np.cos(th2v[i])*th2p,
     -r2*np.sin(th2v[i])*th2p,
     r2*np.cos(th2v[i])*th2p]

    xv = np.linalg.solve(am,bv)
    th3pv[i] = xv[0]
    r3pv[i] = xv[1]
    th5pv[i] = xv[2]
    r6xpv[i] = xv[3]
    
    
    # Acceleration

    bv = [r2*np.cos(th2v[i])*th2p**2 + r4*np.cos(th3v[i])*th3pv[i]**2 + r5*np.cos(th5v[i])*th5pv[i]**2,
      r2*np.sin(th2v[i])*th2p**2 + r4*np.sin(th3v[i])*th3pv[i]**2 + r5*np.sin(th5v[i])*th5pv[i]**2,
     -r2*np.cos(th2v[i])*th2p**2 + r3v[i]*np.cos(th3v[i])*th3pv[i]**2 + 2*r3v[i]*np.sin(th3v[i])*th3pv[i],
     -r2*np.sin(th2v[i])*th2p**2 + r3v[i]*np.sin(th3v[i])*th3pv[i]**2 - 2*r3v[i]*np.cos(th3v[i])*th3pv[i]]

    xv = np.linalg.solve(am,bv) # the result of solve is the vector x
    th3ppv[i] = xv[0]
    r3ppv[i] = xv[1]
    th5ppv[i] = xv[2]
    r6xppv[i] = xv[3]
    

th3pNumv=np.gradient(th3v,th2v)*th2p
r3pNumv=np.gradient(r3v,th2v)*th2p
th5pNumv=np.gradient(th5v,th2v)*th2p
r6xpNumv=np.gradient(r6xv,th2v)*th2p

th3ppNumv=np.gradient(np.gradient(th3v,th2v),th2v)*th2p**2
r3ppNumv=np.gradient(np.gradient(r3v,th2v),th2v)*th2p**2
th5ppNumv=np.gradient(np.gradient(th5v,th2v),th2v)*th2p**2
r6xppNumv=np.gradient(np.gradient(r6xv,th2v),th2v)*th2p**2  
```
