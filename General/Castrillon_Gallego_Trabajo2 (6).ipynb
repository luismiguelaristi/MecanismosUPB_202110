 # Kinematic, velocity and aceleration analysis of a Theo Jansen mechanism
 
 
 
 
 


The Jansen leg mechanism can be used in mobile robotic applications and in gait analysis. (Lalit Patnaik, 2016)

A short description of this mechanisim: 12 links, including a crank link that is actuated by a rotary actuator and a stationary link that is part of the chassis of the legged system and hence does not move with respect to the chassis. All other links and pin joints are unactuated and move because of the motion imparted by the crank. Their positions and orientations are uniquely defined by specifying the crank angle and hence the mechanism has only one degree of freedom (1-DoF) (Lalit Patnaik,2016)

Sanz-Meodio et al. presents a set of rules towards improving energy efficiency in statically stablewalkingrobotsextractedto careful dynamic simulation and analysis of two legged,  mammal and insect configurations for a hexapod robotic platform (Sanz-Merodio,2012)

### Parameters and variables

Main variables: $\theta_2$

Secondary variables: $\theta_3, \theta_4,\theta_5, \theta_6,\theta_7, \theta_8$


## Vector Equations 


$\overrightarrow{r_{1x}}+\overrightarrow{r_{1y}}+\overrightarrow{r_{2}}-\overrightarrow{r_{3}}-\overrightarrow{r_{4b}} = 0$ (2)

$-\overrightarrow{r_{3}}-\overrightarrow{r_{4b}}-\overrightarrow{r_{7}}+\overrightarrow{r_{8}}= 0$ (3)

$-\overrightarrow{r_{3}}-\overrightarrow{r_{4b}}+\overrightarrow{r_{4a}}-\overrightarrow{r_{5}}+\overrightarrow{r_{6a}}+\overrightarrow{r_{8}}= 0$ (4)




## Scalar Equations

$r_{1x}\cdot cos(180)+r_2\cdot cos(\theta_{2})-r_3\cdot cos(\theta_{3})-r_{4b}\cdot cos(\theta_{4}+\theta_{4b})=0$, (5)

$r_{1y}+r_2\cdot sin(\theta_{2})-r_3\cdot sin(\theta_{3})-r_{4b}\cdot sin(\theta_{4}+\theta_{4b})=0$, (6)

$-r_3\cdot cos(\theta_{3})-r_{4b}\cdot cos(\theta_{4}+\theta_{4b})-r_{7}\cdot cos(\theta_{7})+r_{8}\cdot cos(\theta_{8})=0$ (7)

$-r_3\cdot sin(\theta_{3})-r_{4b}\cdot sin(\theta_{4}+\theta_{4b})-r_{7}\cdot sin(\theta_{7})+r_{8}\cdot sin(\theta_{8})=0$ (8)

$-r_3\cdot cos(\theta_{3})-r_{4b}\cdot cos(\theta_{4}+\theta_{4b})+r_{4a}\cdot cos(\theta_{4})-r_{5}\cdot cos(\theta_{5})+r_{6a}\cdot cos(\theta_{6})+r_{8}\cdot cos(\theta_{8})=0$ (9)

$-r_3\cdot sin(\theta_{3})-r_{4b}\cdot sin(\theta_{4}+\theta_{4b})+r_{4a}\cdot sin(\theta_{4})-r_{5}\cdot sin(\theta_{5})+r_{6a}\cdot sin(\theta_{6})+r_{8}\cdot sin(\theta_{8})=0$ (10)

```python
from scipy.optimize import fsolve 
import matplotlib.pyplot as plt 

def theojansen (ini,para):
# Initial conditions 
 th2 = para[0]

## Parameters 

 r1x = para[1]
 r1y = para[2]
 r2 = para[3]
 r3 = para[4]
 r4b = para[5]
 th4b = para[6]
 r4a = para[7]
 r4c = para[8]
 r5 = para[9]
 r6a = para[10]
 r6b = para[11]
 r6c = para[12]
 r7 = para[13]
 r8 = para[14]

##numpy arrays 
 th3=ini[0]
 th4=ini[1]
 th5=ini[2]
 th6=ini[3]
 th7=ini[4]
 th8=ini[5]

 Y= np.zeros(6)
 Y[0] = r1x*np.cos(np.deg2rad(180)) + r2*np.cos(th2) - r3*np.cos(th3) - r4b*np.cos(th4+th4b) + r4a*np.cos(th4) - r5*np.cos(th5) + r6a*np.cos(th6) + r7*np.cos(th7);
 Y[1] = r1y + r2*np.sin(th2) - r3*np.sin(th3) - r4b*np.sin(th4+th4b) + r4a*np.sin(th4) - r5*np.sin(th5) + r6a*np.sin(th6) + r7*np.sin(th7);
 Y[2] = - r3*np.cos(th3) - r4b*np.cos(th4+th4b) - r7*np.cos(th7)+ r8*np.cos(th8);
 Y[3] = - r3*np.sin(th3) - r4b*np.sin(th4+th4b) - r7*np.sin(th7)+ r8*np.sin(th8);
 Y[4] = - r3*np.cos(th3) - r4b*np.cos(th4+th4b) + r4a*np.cos(th4) - r5*np.cos(th5) + r6a*np.cos(th6) + r8*np.cos(th8); 
 Y[5] = - r3*np.sin(th3) - r4b*np.sin(th4+th4b) + r4a*np.sin(th4) - r5*np.sin(th5) + r6a*np.sin(th6) + r8*np.sin(th8); 
 return Y
pass
```

```python
r1x = 380
r1y = 80
r2 = 115
r3 = 500
r4b = 415
th4b = np.deg2rad(72.82)
r4a = 410
r4c = 490
r5 = 394
r6a = 367
r6b = 657
r6c = 490
r7 = 393
r8 = 619

th2 = np.deg2rad(180)

params = [th2,r1x,r1y,r2,r3,r4b,th4b,r4a,r4c,r5,r6a,r6b,r6c,r7,r8]

th3in=np.deg2rad(219.78)
th4in=np.deg2rad(32.61)
th5in=np.deg2rad(66.89)
th6in=np.deg2rad(221.31)
th7in=np.deg2rad(77.52)
th8in=np.deg2rad(131.49)

x0 = [th3in,th4in,th5in,th6in,th7in,th8in]
print(theojansen(x0,params))
```

```python
q,info,exitflag,message = fsolve(theojansen,x0,args = params,full_output = True)
print("solution = ",np.rad2deg(q),"\n",exitflag,"\n",message)
th3 = q[0]
th4 = q[1]
th5 = q[2]
th6 = q[3]
th7 = q[4]
th8 = q[5]
```

# Velocity and aceleration analysis.



We will apply the following equations to model the velocity of each rigid body of the mechanism: 

$\overrightarrow{V_D} = 0 $, 

$\overrightarrow{V_C} =\overrightarrow{\omega}\times \overrightarrow{r_{DC}}$,

$\overrightarrow{V_B} = \overrightarrow{V_C}+\overrightarrow{\omega_4}\times \overrightarrow{r_{4C}}$, 

$\overrightarrow{V_G} = 0$,

$\overrightarrow{V_E} = \overrightarrow{V_C}+\overrightarrow{\omega_6}\times \overrightarrow{r_{6B}}$,

Where $\overrightarrow{\omega}$ is the angular velocity vector.

The value $\overrightarrow{\omega}$,since the scalar components of this vector are:

$\overrightarrow{\omega} = \begin{bmatrix}
0 & 0 & \dot{\theta} 
\end{bmatrix}$ 

Where $\dot{\theta}$ is the magnitude of the angular velocity. This quantity is obtained by deriving the equations of scalar position with respect to time.

```python
import sympy as sym
from sympy.physics.mechanics import dynamicsymbols, init_vprinting

sym.init_printing(use_unicode=False, wrap_line=True)
init_vprinting()
```
